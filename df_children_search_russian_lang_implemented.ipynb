```python
import pandas as pd
import numpy as np
import re
```

```python
#import minjust2018.cvs
df_minjust = pd.read_csv('data/Minjust2018.csv', 
                 header=None, 
                 usecols=[4, 27, 28, 33])
df_minjust.fillna('', inplace=True)
df_minjust.rename(columns={4: 'company_name', 
                           27: 'director', 
                           28: 'activity', 
                           33: 'founders'},
                  inplace = True)
```

```python
#import declaration.csv
df_decl = pd.read_csv('data/declaration.csv', 
                      header=None, 
                      usecols=[1, 2, 3, 4, 5])
df_decl.fillna('', inplace=True)
df_decl.rename(columns={1: 'surname', 
                        2: 'name', 
                        3: 'patronymic', 
                        4: 'workplace',
                        5: 'position'},
                  inplace = True)
```

```python
columns = ['company_name', 'director', 'activity', 'founders', 'surname', 'name', 'patronymic', 'workplace', 'position']
result_child = pd.DataFrame(columns=columns)
result_bro_sis = pd.DataFrame(columns=columns)
```

```python
#Searching for children of ministers
for index, row in df_decl.loc[994:1064].iterrows():
    search = search_df(df_minjust, 'founders', child_regex(row['surname'], row['name'])).copy()
    for column in df_decl.columns:
        search[column] = row[column]
    result_child = result_child.append(search, ignore_index=True)
    
```

```python
#Searching for brothers and sisters of ministers
for index, row in df_decl.loc[994:1064].iterrows():
    search = search_df(df_minjust, 'founders', bro_sis_regex(row['surname'], row['patronymic'])).copy()
    for column in df_decl.columns:
        search[column] = row[column]
    result_bro_sis = result_bro_sis.append(search, ignore_index=True)
```

```python
#export
result_bro_sis.to_csv('min_bro_sis_Edil.csv', index=False)
result_bro_sis.to_pickle('min_bro_sis_Edil.pkl')
```

```python
def search_df(df, column, pattern):
    return df.loc[df[column].str.contains(pattern)]
    
def child_regex(surname, name):
    pattern = re.compile(surname +
                         r'а?\s+\w+\s+' + # First name
                         remove_vowels(name) + # Patronymic
                         r'|' + 
                         r'\w+\s+' + # First name
                         remove_vowels(name) + r'\w+\s+' + # Patronymic
                         surname,
                         re.IGNORECASE)
    return pattern

def remove_vowels(name):
    while True:
        if name[-1] not in ['а', 'о', 'и', 'е', 'ё', 'э', 'ы', 'у', 'ю', 'я', 'й']:
            break
        else:
            name = name[: -1]
    return name

def remove_ends_bro(patronymic):
    while True:
        if patronymic[-3:] not in ['вич', 'вна']:
            break
        else:
            patronymic = patronymic[: -3]
    return patronymic


def bro_sis_regex(surname, patronymic):
    pattern = re.compile(surname +
                         r'а?\s+\w+\s+' + # First name
                         remove_ends_bro(patronymic) + # Patronymic
                         r'|' + 
                         r'\w+\s+' + # First name
                         remove_ends_bro(patronymic) + r'\w+\s+' + # Patronymic
                         surname,
                         re.IGNORECASE)
    return pattern
```
